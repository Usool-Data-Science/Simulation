# A SIMULATION OF CALL (REQUEST AND RESPONSE) IN A CUSTOMER SERVICE CENTER

## STATEMENT OF PROBLEM

### Statements
Assume a call center where two staff members, named Zain and Ahmed, take care of customers to give service. 

And we have the following possible roles:

Rule A:
•	If both are idle, a call is assigned to Zain. 
•	If Ahmed is busy, a call is assigned to Ahmed.
•	If both are busy, a call waits in the queue.

Rule B:
•	If both are idle, a call is assigned to Ahmed. 
•	If Ahmed is busy, a call is assigned to Zain.
•	If both are busy, a call waits in the queue.

Rule C:
•	If both are idle, a call is assigned randomly to Ahmed or Zain. 
•	If Ahmed is busy, a call is assigned to Zain.
•	If Zain is busy, a call is assigned to Ahmed.
•	If both are busy, a call waits in the queue.


Interarrival distribution of calls for technical support
Time between Arrivals [Minute]	Probability	Cumulative Probability
1	0.1	0.1
2	0.5	0.6
3	0.3	0.9
4	0.1	1
		
Service time distribution of Ahmed
Service Time [Minute]	Probability	Cumulative Probability
2	0.4	0.4
3	0.3	0.7
4	0.2	0.9
5	0.1	1
		
Service time distribution of Zain
Service Time [Minute]	Probability	Cumulative Probability
3	0.2	0.2
4	0.4	0.6
5	0.2	0.8
6	0.2	1


### Problem question

a.	Average waiting time
b.	Probability that a caller must wait
c.	Proportion of Ahmed's idle time.
d.	Proportion of Zain’s idle time
e.	Average service time
f.	Average time between arrivals
g.	Average waiting time of those who wait
h.	Average time a caller spends in system
3.	Run the simulation for 400 time for 100 calls and drew a histogram for call average caller delay (for the three rules).
4.	Discuss the finding in task 2 and 3, and determine which rules have the best performance.


```python
import random
```

```python
def simulate_call_center(num_calls, rule):
    zain_busy = False
    ahmed_busy = False
    zain_finish_time = 0
    ahmed_finish_time = 0
    queue = 0
    time = 0
    wait_time = 0
    total_wait_time = 0
    total_service_time = 0
    total_time_in_system = 0
    total_idle_time_ahmed = 0
    total_idle_time_zain = 0
    calls_waiting = 0
    calls_served = 0    
    for i in range(num_calls):
        interarrival_time = random.choices([1, 2, 3, 4], cum_weights=[0.1, 0.6, 0.9, 1])[0]
        time += interarrival_time
        if rule == "A":
            if not zain_busy and not ahmed_busy:
                zain_busy = True
                zain_finish_time = time + random.choices([3, 4, 5, 6], cum_weights=[0.2, 0.6, 0.8, 1])[0]
                total_service_time += zain_finish_time - time
                calls_served += 1
            elif ahmed_busy:
                ahmed_finish_time = time + random.choices([2, 3, 4, 5], cum_weights=[0.4, 0.7, 0.9, 1])[0]
                total_service_time += ahmed_finish_time - time
                calls_served += 1
            else:
                queue += 1
                wait_time = time
                total_wait_time += wait_time
                calls_waiting += 1
        elif rule == "B":
            if not zain_busy and not ahmed_busy:
                ahmed_busy = True
                ahmed_finish_time = time + random.choices([2, 3, 4, 5], cum_weights=[0.4, 0.7, 0.9, 1])[0]
                total_service_time += ahmed_finish_time - time
                calls_served += 1
            elif ahmed_busy:
                zain_busy = True
                zain_finish_time = time + random.choices([3, 4, 5, 6], cum_weights=[0.2, 0.6, 0.8, 1])[0]
                total_service_time += zain_finish_time - time
                calls_served += 1
            else:
                queue += 1
                wait_time = time
                total_wait_time += wait_time
                calls_waiting += 1
        elif rule == "C":
            if not zain_busy and not ahmed_busy:
                if random.random() < 0.5:
                    ahmed_busy = True
                    ahmed_finish_time = time + random.choices([2, 3, 4, 5], cum_weights=[0.4, 0.7, 0.9, 1])[0]
                    total_service_time += ahmed_finish_time - time
                    calls_served += 1
                else:
                    zain_busy = True
                    zain_finish_time = time + random.choices([3, 4, 5, 6], cum_weights=[0.2, 0.6, 0.8, 1])[0]
                    total_service_time += zain_finish_time - time
                    calls_served += 1
            elif ahmed_busy:
                zain_busy = True
                zain_finish_time = time + random.choices([3, 4, 5, 6], cum_weights=[0.2, 0.6, 0.8, 1])[0]
                total_service_time += zain_finish_time - time
                calls_served += 1
            elif zain_busy:
                queue += 1
                wait_time = time
                total_wait_time += wait_time
                calls_waiting += 1
        # update the idle time for each agent
        if not zain_busy:
            total_idle_time_zain += time - zain_finish_time
        if not ahmed_busy:
            total_idle_time_ahmed += time - ahmed_finish_time
        # update the time in system for each call
        total_time_in_system += wait_time + (time - wait_time)
    return num_calls, total_wait_time, calls_waiting, total_idle_time_ahmed,time,total_idle_time_zain,total_service_time,calls_served,total_time_in_system

```

```python
for rule in ["A", "B", "C"]:
    num_calls, total_wait_time, calls_waiting, total_idle_time_ahmed,time,total_idle_time_zain,total_service_time,calls_served,total_time_in_system = simulate_call_center(1000, rule)
    print("Rule: ", rule, '==============================')
    #a. Average waiting time:
    # The average waiting time is obtained by dividing the total wait time by the number of calls that had to wait in the queue.
    if calls_waiting == 0:
        print("Average waiting time is undefined as there are no calls waiting")
    else:
        avg_wait_time = total_wait_time / calls_waiting
        print("Average waiting time is: ", avg_wait_time)
    
    #b. Probability that a caller must wait:
    # The probability that a caller must wait is obtained by dividing the number of calls that had to wait in the queue by the total number of calls.
    prob_wait = calls_waiting / num_calls
    print('Probability that a caller must wait is: ', prob_wait)
    
    #c. Proportion of Ahmed's idle time:
    # The proportion of Ahmed's idle time is obtained by dividing the total idle time of Ahmed by the total simulation time.
    proportion_ahmed_idle = total_idle_time_ahmed / time
    print("Proportion of Ahmed's idle time is: ", proportion_ahmed_idle)
    
    #d. Proportion of Zain’s idle time:
    # The proportion of Zain's idle time is obtained by dividing the total idle time of Zain by the total simulation time.
    proportion_zain_idle = total_idle_time_zain / time
    print("Proportion of zain's idle time is:", proportion_zain_idle)
    
    #e. Average service time:
    # The average service time is obtained by dividing the total service time by the number of calls served.
    if calls_served == 0:
        print("The average service time is undefined as there are no calls served")
    else:
        avg_service_time = total_service_time / calls_served
        print("The average service time is:", avg_service_time)
    #f. Average time between arrivals:
    # The average time between arrivals can be obtained by dividing the total time between the arrivals by the number of calls.
    if num_calls == 0:
        print("The average time between arrivals is undefined as there are no calls made")
    else:
        avg_interarrival_time = time / num_calls
        print("The average time between arrivals is:", avg_interarrival_time)
        
    #g. Average waiting time of those who wait:
    # The average waiting time of those who wait can be obtained by dividing the total wait time by the number of calls that had to wait in the queue.
    if calls_waiting == 0:
        print("The average waiting time of those who wait is undefined as there are no calls waiting")
    else:
        avg_wait_time_who_wait = total_wait_time / calls_waiting
        print("The average waiting time of those who wait is: ", avg_wait_time_who_wait)
    
    #h. Average time a caller spends in system:
    # The average time a caller spends in the system can be obtained by dividing the total time a caller spends in the system by the number of calls.
    if num_calls == 0:
        print("The average time a caller spends in the system is undefined as there are no calls made")
    else:
        avg_time_in_system = total_time_in_system / num_calls
        print("The average time a caller spends in the system is: ", avg_time_in_system)
```
